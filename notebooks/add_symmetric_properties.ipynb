```python
import pandas as pd
from qwikidata.entity import WikidataItem, WikidataLexeme, WikidataProperty
from qwikidata.linked_data_interface import get_entity_dict_from_api
```

```python
symmetric_relations = pd.read_csv('../data/knowledge/symmetric_multilingual.csv')
symmetric_relations
```

```python
ru = []
zh = []
ja = []

for pid in list(symmetric_relations['id']):
    p_dict = get_entity_dict_from_api(pid)
    p = dict((k, p_dict[k]) for k in ('id', 'type', 'labels'))
    if 'ru' in p['labels']:
        ru.append(p['labels']['ru']['value'])
    else:
        ru.append(None)
    
    if 'zh' in p['labels']:
        zh.append(p['labels']['zh']['value'])
    else:
        zh.append(None)
    
    if 'ja' in p['labels']:
        ja.append(p['labels']['ja']['value'])
    else:
        ja.append(None)
```

```python
symmetric_relations['zh'] = zh
```

```python
symmetric_relations
```

```python
symmetric_relations.to_csv('../data/knowledge/symmetric_multilingual.csv')
```

Now take every relations dataset and add the relations if they exist in all languages

```python
# Load relations
path = '../data/knowledge/zh_ja_relations.csv'
df = pd.read_csv(path)
df
```

```python
# Add relations
for index, relation in symmetric_relations.iterrows():
    if relation['zh'] is not None and relation['ja'] is not None:
        # Add row to df
        rel_dict = dict((k, relation[k]) for k in ('id', 'zh', 'ja'))
        df = df.append(rel_dict, ignore_index=True)
        
```

```python
df
```

```python
# Save again
df.to_csv(path, index=False)
```
