```python
import pandas as pd
from tqdm import tqdm
import translators as ts
from googletrans import Translator
from collections import defaultdict

translator = Translator()
```

```python
import time
```

```python
relations = pd.read_csv('../data/relations/relations_en_de_es_fr.csv')
relations
```

```python
languages = ['en', 'de', 'fr', 'es']
```

```python
translate_dict = defaultdict(list)

for target_lang in languages:
    
    no = translator.translate('not', src='en', dest=target_lang).text

    # Iterate over relations
    for idx, relation in tqdm(relations.iterrows(), total=relations.shape[0]):

        translate_dict[target_lang + '_negated'].append(f'{no} {relation[target_lang]}')
```

```python
for lang in languages:
    relations[lang + '_negated'] = translate_dict[lang + '_negated']
```

```python
relations.to_csv('../data/relations/negation/negation_general_en_de_es_fr.csv')
```
