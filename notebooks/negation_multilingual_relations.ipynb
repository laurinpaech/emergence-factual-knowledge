```python
import pandas as pd
from tqdm import tqdm
import translators as ts
from googletrans import Translator
from collections import defaultdict

translator = Translator()
```

```python
import time
```

```python
relations = pd.read_csv('../data/relations/negation/negation_en.csv')
relations
```

```python
languages = ['de', 'fr', 'es']
```

```python
translate_dict = defaultdict(list)

for target_lang in languages:

    # Iterate over relations
    for idx, relation in tqdm(relations.iterrows(), total=relations.shape[0]):

        translation = translator.translate(relation['en'], src='en', dest=target_lang).text
        translation_negated = translator.translate(relation['en_negated'], src='en', dest=target_lang).text
        time.sleep(1)

        # If empty, use None instead
        if not translation:
            translation = None

        translate_dict[target_lang].append(translation)
        translate_dict[target_lang + '_negated'].append(translation_negated)
```

```python
for lang in languages:
    relations[lang] = translate_dict[lang]
    relations[lang + '_negated'] = translate_dict[lang + '_negated']
```
